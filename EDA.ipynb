```python
# Suppress the deprecation warnings in the output for cleaner visualization code
import warnings
warnings.filterwarnings("ignore")

data = pd.read_csv('data/raw/BankChurners.csv')
```

```python
# --- 2. 결측치 분석 (Missing Value Analysis) ---
print("--- [2. 결측치 분석 결과] ---")
missing_info = pd.DataFrame({
    'Missing Count': data.isnull().sum(),
    'Missing Percentage': (data.isnull().sum() / len(data)) * 100
})

# Filter out columns with 0 missing values for a concise report
missing_info = missing_info[missing_info['Missing Count'] > 0]

if missing_info.empty:
    print("모든 열에 명시적인 결측치 (NaN)는 없습니다.")
else:
    print(missing_info.to_markdown(numalign="left", stralign="left"))
```

```python
# --- 1. 이탈률 분석 및 그래프 생성 준비 (Churn Analysis & Prep) ---

# Attrition_Flag를 이진 변수(1: 이탈, 0: 유지)로 변환
data['Churn'] = data['Attrition_Flag'].apply(lambda x: 1 if x == 'Attrited Customer' else 0)

# 전체 평균 이탈률 계산
overall_churn_rate = data['Churn'].mean()
print(f"\n--- [1. 전체 평균 이탈률] ---")
print(f"전체 평균 이탈률: {overall_churn_rate:.4f} ({overall_churn_rate * 100:.2f}%)")

# 이탈률 분석을 위한 2개 열 선택: 'Income_Category' (소득 범주) 및 'Churn'
grouped_churn = data.groupby('Income_Category')['Churn'].mean().reset_index()
grouped_churn = grouped_churn.sort_values(by='Churn', ascending=False)

# --- 1. 다양한 그래프 코드 생성 (Visualization Code Generation) ---
print("\n--- [1. 다양한 이탈률 평균 그래프 코드 생성] ---")

# 1-A. 소득 범주별 평균 이탈률 (막대 그래프: Bar Chart)
plt.figure(figsize=(10, 6))
sns.barplot(x='Income_Category', y='Churn', data=grouped_churn, palette='viridis')
plt.title('소득 범주별 고객 평균 이탈률', fontsize=15)
plt.xlabel('소득 범주', fontsize=12)
plt.ylabel('평균 이탈률 (Churn Rate)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('churn_rate_by_income.png')
print("-> 소득 범주별 이탈률 막대 그래프: 'churn_rate_by_income.png' 저장")
plt.close()

# 1-B. 고객 연령별 이탈 여부 분포 (밀도 그래프: Distribution Plot)
plt.figure(figsize=(10, 6))
sns.histplot(
    data=data, 
    x='Customer_Age', 
    hue='Attrition_Flag', 
    kde=True, 
    palette={'Existing Customer': 'skyblue', 'Attrited Customer': 'salmon'},
    stat="density", 
    common_norm=False,
    linewidth=0.5
)
plt.title('Density Distribution of Customer Churn by Age', fontsize=15)
plt.xlabel('Customer Age', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.legend(title='이탈 여부', labels=['이탈 고객', '기존 고객'])
plt.tight_layout()
plt.savefig('churn_distribution_by_age.png')
print("-> 고객 연령별 이탈 여부 밀도 그래프: 'churn_distribution_by_age.png' 저장")
plt.show()
```

```python
# Unknown 값 확인. 

unknown_counts = {}
for col in ['Education_Level', 'Marital_Status', 'Income_Category']:
    if (data[col] == 'Unknown').any():
        count = (data[col] == 'Unknown').sum()
        percentage = (count / len(data)) * 100
        unknown_counts[col] = (count, percentage)

print("\n--- [3. Feature Engineering 분석 정보] ---")
if unknown_counts:
    print("특정 범주형 변수에 'Unknown' 값이 포함되어 있습니다 (결측치와 유사하게 처리 가능).")
    for col, (count, percentage) in unknown_counts.items():
        print(f"  - {col}: {count}개 ({percentage:.2f}%)")
```
